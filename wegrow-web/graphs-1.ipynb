```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from json import loads

def get_points(x, y):
    a = 'old-stuff/fishy-wifi-master/wegrow-web/wegrow-data1/wegrow-data/WaterElf-10865861/2016-06-23_WaterElf-10865861.txt'
    data = {}
    with open(a,'r') as f:
        for line in f:
            new_dict = loads(line.strip())
            for k in new_dict:
                if k in data:
                    data[k].append(float(new_dict[k]))
                else:
                    data[k] = [float(new_dict[k])]

    array = np.array([data[x], data[y]])
    plt.plot(array[0], array[1],label=y)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.legend()
    plt.show()
    
get_points('timestamp', 'airTemp')
get_points('timestamp', 'humidity')
```
